```python
import torch
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create some toy data

```python
num_data = 100
```

```python
x_train = np.array(range(num_data), dtype=np.float32)
```

```python
coef = 3
bias = 4
noise = num_data*(np.random.rand(num_data)-0.5)
y_train = x_train*coef+bias+noise
x_train=x_train.reshape(num_data,1)
y_train=y_train.astype(np.float32).reshape(num_data,1)
```

```python
plt.plot(x_train,y_train,'v')
```

## Create a linear model

```python
model = torch.nn.Linear(1,1)
```

```python
loss_fn = torch.nn.MSELoss()
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.000001)
```

```python
for epoch in range(1000):
    y_pred = model(torch.from_numpy(x_train))
    loss = loss_fn(y_pred, torch.from_numpy(y_train))
    optimizer.zero_grad()
    loss.backward()    
    optimizer.step()
    if epoch%100==0:
        print(list(model.parameters()))
        plt.plot(x_train,y_train,'v', label='Original Data')
        plt.plot(x_train,y_pred.data.numpy(), label='Fitted Line')
        plt.legend()
        plt.show()
    
```
