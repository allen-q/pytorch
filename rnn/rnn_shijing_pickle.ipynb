```python
import numpy as np
import pickle
```

```python
with open('rnn_shijing.pickle', 'rb') as handle:
    all_vars = pickle.load(handle)
```

```python
def sample(h, seed_ix, n):
  """ 
  sample a sequence of integers from the model 
  h is memory state, seed_ix is seed letter for first time step
  """
  x = np.zeros((vocab_size, 1))
  x[seed_ix] = 1
  ixes = []
  for t in range(n):
    h = np.tanh(np.dot(Wxh, x) + np.dot(Whh, h) + bh)
    y = np.dot(Why, h) + by
    p = np.exp(y) / np.sum(np.exp(y))
    ix = np.random.choice(range(vocab_size), p=p.ravel())
    x = np.zeros((vocab_size, 1))
    x[ix] = 1
    ixes.append(ix)
  return ixes
```

```python
vocab_size = all_vars['vocab_size']
Wxh = all_vars['Wxh']
Whh = all_vars['Whh']
bh = all_vars['bh']
Why = all_vars['Why']
by = all_vars['by']
ix_to_char = all_vars['ix_to_char']
char_to_ix = all_vars['char_to_ix']
hprev = all_vars['hprev']
inputs = all_vars['inputs']
```

```python
print(''.join(ix_to_char[ix] for ix in sample(hprev, inputs[0], 5000)))
```
