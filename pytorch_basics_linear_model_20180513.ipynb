```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_data = 10
x_train = np.array(range(num_data))
```

```python
x_train
```

```python
coef = 3
bias = 5
noise = 5*(np.random.rand(10)-0.5)
```

```python
y_train = x_train*coef+bias+noise
```

```python
x_train = x_train.reshape(num_data,1).astype(np.float32)
y_train = y_train.reshape(num_data,1).astype(np.float32)
```

```python
plt.plot(x_train, y_train, 'v')
```

```python
model = torch.nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001)
```

```python
    y_pred = model(torch.from_numpy(x_train))
    optimizer.zero_grad()    
    loss = loss_fn(y_pred, torch.from_numpy(y_train))
    loss.backward()
```

```python
for epoch in range(20000):
    y_pred = model(torch.from_numpy(x_train))
    optimizer.zero_grad()    
    loss = loss_fn(y_pred, torch.from_numpy(y_train))
    loss.backward()
    optimizer.step()
    
    if epoch%100==0:
        print('loss at epoch {} is:{}'.format(epoch, loss))
        print('coef:{}. bias:{}'.format(*[e.data for e in model.parameters()]))
        plt.plot(x_train, y_train, 'v', label='Original Data')
        plt.plot(x_train, y_pred.data.numpy(), label='Fitted Line')
        plt.legend()
        plt.show()
```
