```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
% matplotlib inline
```

## Create some toy data

```python
iris = datasets.load_iris()
```

```python
x_train = iris.data[:,0:2].astype(np.float32)
```

```python
y_train = iris.target.astype(np.int64)
```

```python
plt.scatter(x_train[:,0], x_train[:,1],s=100,c=y_train,edgecolors='k',cmap=plt.cm.Paired)
```

## Create a Linear Model

```python
model = nn.Linear(2,3)
```

## Define Loss Functions and Optimizer

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

## Sanity Check

### Manually work out the loss

```python
y_pred_score = model(torch.from_numpy(x_train))
```

```python
# exponentiate the score to get the prob for each class
y_pred_prob = np.exp(y_pred_score.data.numpy())
```

```python
# normalize the prob so they sum up to 1
y_pred_prob = y_pred_prob/y_pred_prob.sum(1)[:,None]
```

### Calculate the loss using the formula below:

$$ L = { \frac{1}{m}\sum_i -\log\left(\frac{e^{f_{y_i}}}{ \sum_j e^{f_j} }\right) }  $$

```python
-np.mean(np.log(np.choose(y_train, y_pred_prob.T)))
```

### The loss is the same calculated by CrossEntropyLoss function

```python
loss = loss_fn(y_pred_score, torch.from_numpy(y_train))
```

```python
loss
```

## Train the model

```python
for epoch in range(1, 30001):
    y_pred_score = model(torch.from_numpy(x_train))
    optimizer.zero_grad()
    loss = loss_fn(y_pred_score, torch.from_numpy(y_train))
    loss.backward()
    optimizer.step()
    
    if epoch%1000 == 0:
        print(f'Loss:{loss} at epoch {epoch}')
```

```python
def predict(m, X):
    if isinstance(X, np.ndarray):
        X = torch.from_numpy(X.astype(np.float32))
    return torch.max(m(X),1)[1]
```

## Plot the decision boundary

```python
x1_min, x2_min = x_train.min(0)-0.5
```

```python
x1_max, x2_max = x_train.max(0)+0.5
```

```python
x1, x2 = np.meshgrid(np.arange(x1_min, x1_max,0.01), np.arange(x2_min, x2_max, 0.01))
```

```python
z = predict(model, np.c_[x1.ravel(), x2.ravel()])
```

```python
plt.figure(figsize=(10,6))
plt.pcolormesh(x1, x2, z.reshape(x1.shape), cmap=plt.cm.Paired)
plt.scatter(x_train[:,0], x_train[:,1], c=y_train, edgecolors='k', s=150, cmap=plt.cm.Paired)
plt.show()
```
