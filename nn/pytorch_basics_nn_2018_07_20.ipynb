```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
r = np.tile(np.r_[0:1:100j],3)
t = np.r_[0:np.pi*4:300j] + np.random.random(300)
X = np.c_[r*np.cos(t), r*np.sin(t)]
y = np.repeat(np.arange(3), 100)
plt.scatter(*X.T,c=y, cmap=plt.cm.Paired)
```

```python
class NeuralNet(nn.Module):
    def __init__(self, num_hidden):
        super().__init__()
        self.fc1 = nn.Linear(2,num_hidden)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(num_hidden,3)
        
    def forward(self, X):
        fc1_out = self.relu1(self.fc1(X))
        fc2_out = self.fc2(fc1_out)
        
        return fc2_out
    
    def plot_decision_boundary(self, X, y):
        grid = np.meshgrid(np.r_[-1.5:1.5:1000j],np.r_[-1.5:1.5:1000j])
        x_grid = torch.from_numpy(np.r_[grid].reshape(2,-1).T).float()
        y_grid_score = self.forward(x_grid)
        y_grid = y_grid_score.argmax(1).detach().numpy()
        plt.pcolormesh(*grid, y_grid.reshape(grid[0].shape[0],-1), cmap=plt.cm.Paired)
        plt.scatter(*X.T, c=y, cmap=plt.cm.Paired, edgecolors='k')
        plt.show()
```

```python
net1 = NeuralNet(32)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net1.parameters(), lr=0.001)
```

```python
X_tensor = torch.from_numpy(X).float()
y_tensor = torch.from_numpy(y).long()
```

```python
for iter in range(10000):
    optimizer.zero_grad()
    y_score = net1(X_tensor)
    loss = loss_fn(y_score, y_tensor)
    loss.backward()
    optimizer.step()
    
    if iter%1000 == 0:
        print(f'Loss is {loss} at iteration {iter}')
        net1.plot_decision_boundary(X, y)
```
